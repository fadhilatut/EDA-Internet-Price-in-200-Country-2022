### Preprocessing

Drop Kolom yang tidak diperlukan

```python
df = df.drop(columns=['Cheapest 1GB for 30 days (USD)','Most expensive 1GB (USD)'])
df
```

Cek kolom yang mempunyai nilai NaN

```python
df.isnull().sum().sort_values(ascending=False)
```

Drop Null Data

```python
df = df[pd.notnull(df['NO. OF Internet Plans '])]
df.isnull().sum().sort_values(ascending=False)
```

# Mengubah Type Data

Menghilangkan koma (,) pada kolom population dan internet user

```python
df["Population"] = [float(str(i).replace(",", "")) for i in df["Population"]]
df["Internet users"] = [float(str(i).replace(",", "")) for i in df["Internet users"]]
```

Mengganti special value (NO PACKAGES) pada 2 kolom menjadi 0

```python
df['Average price of 1GB (USD  at the start of 2021)'] = df['Average price of 1GB (USD  at the start of 2021)'].replace(['NO PACKAGES'],0)
df['Average price of 1GB (USD – at start of 2020)'] = df['Average price of 1GB (USD – at start of 2020)'].replace(['NO PACKAGES'],0)
```

Mengubah tipe data pada kolom yang belum sesuai

```python
df['Average price of 1GB (USD  at the start of 2021)']=df['Average price of 1GB (USD  at the start of 2021)'].astype('float')
df['Average price of 1GB (USD – at start of 2020)'] = df['Average price of 1GB (USD – at start of 2020)'].astype('float')
```

```python
df['Average price of 1GB (USD)']= df['Average price of 1GB (USD)'].astype('float')
df.info()
```

Melihat data duplikat pada kolom Country

```python
df = df.drop_duplicates(subset='Country')
```

Melihat Outlier data

```python
vars = ['Avg \n(Mbit/s)Ookla','Internet users','Population']
fig = make_subplots(rows=1, cols=len(vars))
for i, var in enumerate(vars):
  fig.add_trace(
      go.Box(y=df[var],name=var),
      row=1,col=i+1
  )
fig.update_traces()
```

Mengisi nilai NaN

```python
#Replcae NaN value
nan_replace = {'Internet users':df['Internet users'].median(),'Population':df['Population'].median(),\
               'Avg \n(Mbit/s)Ookla':df['Avg \n(Mbit/s)Ookla'].median()}
df = df.fillna(nan_replace)
```
