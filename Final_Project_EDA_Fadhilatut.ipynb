### Preprocessing

First drop unnecessary column, i decided to drop these 2 column because i will only use the average price.

```python
df = df.drop(columns=['Cheapest 1GB for 30 days (USD)','Most expensive 1GB (USD)'])
df
```

Then i checked the NaN values for each column

```python
df.isnull().sum().sort_values(ascending=False)
```

3 of 12 column has same NaN value, and after look the datas the NaN values were on the same row, so i drop the 12 row, because those rows is contain unavaiable values

```python
df = df[pd.notnull(df['NO. OF Internet Plans '])]
df.isnull().sum().sort_values(ascending=False)
```

# Change Data Type

After i checked the datas in each column,all row in these 2 column were contain numeric values, yet it still labelled as object, because the numeric values contain comma (,) and it counted as string. Therefore i remove all comas in these 2 column

```python
df["Population"] = [float(str(i).replace(",", "")) for i in df["Population"]]
df["Internet users"] = [float(str(i).replace(",", "")) for i in df["Internet users"]]
```

In these 2 columns, i found out they had special values in both of them. it was the "NO PACKAGES" values. then i replace those value with 0, because No Pacakages can also mean 0 

```python
df['Average price of 1GB (USD  at the start of 2021)'] = df['Average price of 1GB (USD  at the start of 2021)'].replace(['NO PACKAGES'],0)
df['Average price of 1GB (USD – at start of 2020)'] = df['Average price of 1GB (USD – at start of 2020)'].replace(['NO PACKAGES'],0)
```

After replace the value in those columns to all numeric, i change the column datatype to float64

```python
df['Average price of 1GB (USD  at the start of 2021)']=df['Average price of 1GB (USD  at the start of 2021)'].astype('float')
df['Average price of 1GB (USD – at start of 2020)'] = df['Average price of 1GB (USD – at start of 2020)'].astype('float')
```

```python
df['Average price of 1GB (USD)']= df['Average price of 1GB (USD)'].astype('float')
df.info()
```

After all column datatypes were changed correctly, i check duplicate data in some column. i found duplicate in column Country, and then drop the duplicate

```python
df.shape
```

```python
df.duplicated(subset='Country').sum()
```

```python
df = df.drop_duplicates(subset='Country')
```

Check the duplicated values

```python
df.duplicated(subset='Country').sum()
```

Resetting the index after drop the duplicate value

```python
df = df.reset_index(drop=True)
```

```python
df.info()
```

View the outlier data. the last step of this preprocessing is filling the remaining NaN value, before that i check the data outliers

```python
vars = ['Avg \n(Mbit/s)Ookla','Internet users','Population']
fig = make_subplots(rows=1, cols=len(vars))
for i, var in enumerate(vars):
  fig.add_trace(
      go.Box(y=df[var],name=var),
      row=1,col=i+1
  )
fig.update_traces()
```

Replacing NaN value with median of each column. 

```python
#Replcae NaN value
nan_replace = {'Internet users':df['Internet users'].median(),'Population':df['Population'].median(),\
               'Avg \n(Mbit/s)Ookla':df['Avg \n(Mbit/s)Ookla'].median()}
df = df.fillna(nan_replace)
```

# 5. Price changes in 2020-2021

```python
price = df['Average price of 1GB (USD  at the start of 2021)']-df['Average price of 1GB (USD – at start of 2020)']
df['Price_diff'] = price
df.loc[df['Price_diff']< 0, 'S_Price'] = 'Decrease'
df.loc[df['Price_diff']> 0, 'S_Price'] = 'Increase'
df.loc[df['Price_diff']== 0, 'S_Price'] = 'Fixed'
dfa = df['S_Price'].value_counts()
#px.bar(dfa)
dfa
```

There are 185 country that the average price of internet were decreased in 2020-2021, 43 country that the average price of internet were increased, and 1 country with fixed price

#6. Average price differences in 2020-2021

```python
prc_plus = (df['Price_diff'] / df['Average price of 1GB (USD – at start of 2020)'])*100
prc_min = (df['Price_diff'] / df['Average price of 1GB (USD  at the start of 2021)'])*100
prc_min

df.loc[df['Price_diff']< 0, 'Per_Price_diff (%)'] = prc_plus
df.loc[df['Price_diff']> 0, 'Per_Price_diff (%)'] = prc_min
df.loc[df['Price_diff']== 0, 'Per_Price_diff (%)'] = 0
```
